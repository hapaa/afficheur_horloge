# Conversion chiffre en version V3 pour chaque cadran

```python
#Cadran en Bas Gauche
def number_to_V3_BG (number):
    Angle = np.zeros((11, 11))
    Rayon = np.zeros((11, 11))
    
    for i in range(1,len(Angle)):
        for k in range(1,len(Angle)):
            r = np.sqrt(i*i+(k-len(Angle))*(k-len(Angle)))
            Rayon[i-1][k-1] = int(r)
            Angle[i-1][k-1] = int(np.arcsin(i/r)*180/np.pi/3)

    Value = number

    Temp = np.zeros((15*2,16))

    for i in range(0, len(Angle)) :
        for k in range(0, len(Angle)) :
            alpha = int(Angle[i][k])
            r = int(Rayon[i][k])
            Temp[alpha][15-r] = Value[len(Angle)-1-k][len(Angle)-1-i]

    Temp = Temp.tolist()

    for i in range(0, 15*2):
        for k in range(0,16):
            Temp[i][k]=int(Temp[i][k])
        Temp[i] = ''.join(str(k) for k in Temp[i])
        Temp[i] = int(Temp[i],2)
        
    return Temp

Angle = np.zeros((11, 11))
Rayon = np.zeros((11, 11))


#Cadran en Haut Gauche
def number_to_V3_HG (number):
    Angle = np.zeros((11, 11))
    Rayon = np.zeros((11, 11))

    for i in range(1,len(Angle)):
        for k in range(1,len(Angle)):
            r = np.sqrt((len(Angle)-i)*(len(Angle)-i)+(len(Angle)-k)*(len(Angle)-k))
            Rayon[i-1][k-1] = int(r)
            Angle[i-1][k-1] = int(np.arcsin((len(Angle)-i)/r)*180/np.pi/3)

    Value = number
    
    Temp = np.zeros((15*2,16))

    for i in range(0, len(Angle)) :
        for k in range(0, len(Angle)) :
            alpha = int(Angle[i][k])
            r = int(Rayon[i][k])
            Temp[alpha][15-r] = Value[i][k]

    Temp = Temp.tolist()

    for i in range(0, 15*2):
        for k in range(0,16):
            Temp[i][k]=int(Temp[i][k])
        Temp[i] = ''.join(str(k) for k in Temp[i])
        Temp[i] = int(Temp[i],2)

    return Temp


#Cadran en Haut Droite
def number_to_V3_HD (number):
    Angle = np.zeros((11, 11))
    Rayon = np.zeros((11, 11))

    for i in range(1,len(Angle)):
        for k in range(1,len(Angle)):
            r = np.sqrt((len(Angle)-i)*(len(Angle)-i)+k*k)
            Rayon[i-1][k-1] = int(r)
            Angle[i-1][k-1] = int(np.arcsin((len(Angle)-i)/r)*180/np.pi/3)
    Value = number
    Temp = np.zeros((15*2,16))

    for i in range(0, len(Angle)) :
        for k in range(0, len(Angle)) :
            alpha = int(Angle[i][k])
            r = int(Rayon[i][k])
            Temp[alpha][15-r] = Value[len(Angle)-1-k][len(Angle)-1-i]

    Temp = Temp.tolist()

    for i in range(0, 15*2):
        for k in range(0,16):
            Temp[i][k]=int(Temp[i][k])
        Temp[i] = ''.join(str(k) for k in Temp[i])
        Temp[i] = int(Temp[i],2)

    return(Temp)


#Cadran en Bas Droite
def number_to_V3_BD (number):
    Angle = np.zeros((11, 11))
    Rayon = np.zeros((11, 11))

    for i in range(1,len(Angle)):
        for k in range(1,len(Angle)):
            r = np.sqrt(i*i+k*k)
            Rayon[i-1][k-1] = int(r)
            Angle[i-1][k-1] = int(np.arcsin(i/r)*180/np.pi/3)

    Value = number

    Temp = np.zeros((15*2,16))

    for i in range(0, len(Angle)) :
        for k in range(0, len(Angle)) :
            alpha = int(Angle[i][k])
            r = int(Rayon[i][k])
            Temp[alpha][15-r] = Value[i][k]

    Temp = Temp.tolist()

    for i in range(0, 15*2):
        for k in range(0,16):
            Temp[i][k]=int(Temp[i][k])
        Temp[i] = ''.join(str(k) for k in Temp[i])
        Temp[i] = int(Temp[i],2)

    return(Temp)
```

```python
BG = 1
HG = 2
HD = 3
BD = 4

# Nombre + Cadran Concerne -> nombre en V3
def number_to_V3 (number, cadran):
    if (cadran == BG):
        return number_to_V3_BG (number);
    
    if (cadran == HG):
        return number_to_V3_HG (number);
        
    if (cadran == HD):
        return number_to_V3_HD (number);
    
    if (cadran == BD):
        return number_to_V3_BD (number);
```
